```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#Main dataset 
anime_dataset = pd.DataFrame()

for f in range(139):
  # url of anime website
  url = 'https://zoro.to/completed?page={}'.format(f)

  # getting a webpage 
  webpage = requests.get(url).text

  # applying BeautifulSoup on webpage 
  soup = BeautifulSoup(webpage,'lxml')


  # anime image url list
  anime_img_url = []

  # anime name 
  anime_name = []

  # which one is available sub or dub
  sub_dub = []

  for i in soup.find_all('div',class_='flw-item'):

    # fetching dub or sub

    #checking if ltr class exist or not
    if i.find_all('div',class_='ltr'):
      for j in i.find_all('div',class_='ltr'):
        temp = []
        for k in j.find_all('div',class_='tick-dub'):
          temp.append(k.text)
        sub_dub.append(temp)

    # if ltr class does not exist then put SUB
    else: sub_dub.append('SUB')
    
    # getting name of the anime
    for j in i.find_all('a',class_='dynamic-name'):
      anime_name.append(j.text)
    
    # getting image url
    for j in i.find_all('img',class_='film-poster-img'):
      anime_img_url.append(j['data-src'])

  anime_dis = {'Anime_Name' : anime_name,'Anime_img_url': anime_img_url,'Sub_Dub':sub_dub}
  temp = pd.DataFrame(anime_dis)
  anime_dataset = pd.concat([anime_dataset,temp],ignore_index=True)
```
